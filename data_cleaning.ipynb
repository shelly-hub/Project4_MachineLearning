```python
#Import required packages
import pandas as pd
import numpy as np 
from datetime import datetime
```

### Import Data

```python
#Import data from csv & save to df
data_df = pd.read_csv("canine_cushings_disease_raw.csv")

#Check df
data_df
```

```python
#Check df parameters
data_df.info()
```

### Clean Table (Columns & Rows)

```python
#Set PatientID as index
data_df.index = data_df['PatientID']

#Remove PatientID column
del data_df['PatientID']

#Check df
data_df.head()
```

```python
#Check data types per column
data_df.dtypes
```

### Age

```python
#Change BirthDate to datetime
data_df['BirthDate'] = pd.to_datetime(data_df['BirthDate']).dt.strftime('%Y-%m-%d')

#Change Date first suspected_2 to datetime
data_df['Date first suspected_2'] = pd.to_datetime(data_df['Date first suspected_2']).dt.strftime('%Y-%m-%d')

#Check changes have applied
data_df.head()
```

```python
#Define calculate_age function
def calculate_age(date_str1, date_str2):
    date_format = "%Y-%m-%d"
    date1 = datetime.strptime(date_str1, date_format)
    date2 = datetime.strptime(date_str2, date_format)
    age_in_days = (date2 - date1).days
    age_in_years = age_in_days // 365
    return age_in_years

#Create list to append to
age_data = []

#Loop through column values
for i in range(len(data_df)):

    # Step 1: Provide the two dates
    date_str1 = data_df.iloc[i, 0] # Replace this with the first date in 'YYYY-MM-DD' format
    date_str2 = data_df.iloc[i, 6]  # Replace this with the second date in 'YYYY-MM-DD' format
    
    # Step 3: Calculate the age
    age = calculate_age(date_str1, date_str2)

    # Step 4: Append to list
    age_data.append(age)

#Add Age column to df
data_df['Age'] = age_data

#Check df
data_df.head()
```

### Sex Column

```python
#Create list to append to
sex_data = []

#Loop through column values
for i in range(len(data_df)):

    #Step 1: Save SexSource value
    sex = data_df.iloc[i, 2]

    #Step 2: Split first character and save
    sex_initial = sex[0]
    
    #Step 3: Append to list
    sex_data.append(sex_initial)

#Create Sex column
data_df['Sex'] = sex_data

#Check df
data_df.head()
```

```python
#Remove unwanted columns
data_df.drop(data_df[['SexSource', 'Date confirmed_3', 'Date ruled out_4', 'BirthDate', 'Date first suspected_2', 'ALKP level_28', 'USG level_29', 'BreedVeNomID']], axis=1, inplace=True)

#Check df
data_df.head()
```

### Weight Change

```python
#Summarise Weight Change variable
data_df['Weight_11'].value_counts()
```

```python
#Change 'False' to 'No change'
data_df['Weight_11'] = data_df['Weight_11'].replace('No', "No change")

#Check changes applied
data_df['Weight_11'].value_counts()
```

### Breed Binning

```python
# Choose a cutoff value and create a list of source breeds to be replaced
# use the variable name `breeds_to_replace`
# Need to change value counts to list
breed_counts = data_df['SourceBreed'].value_counts()
breeds_to_replace = data_df['SourceBreed'].value_counts()[breed_counts < 15].index.tolist()

# print(application_types_to_replace)
# Replace in dataframe
for breed in breeds_to_replace:
    data_df['SourceBreed'] = data_df['SourceBreed'].replace(breed,"Others")

# Check to make sure binning was successful
data_df['SourceBreed'].value_counts()
```

### NA Values

```python
#Replace Unknown with NA values
data_df['Increased ALKP_19'] = data_df['Increased ALKP_19'].replace('Unknown', np.nan)
data_df['Increased ALT/GGT/AST_20'] = data_df['Increased ALT/GGT/AST_20'].replace('Unknown', np.nan)
data_df['USG <1.20_21'] = data_df['USG <1.20_21'].replace('Unknown', np.nan)
data_df['Proteinuria_22'] = data_df['Proteinuria_22'].replace('Unknown', np.nan)

#Check df
data_df[['Increased ALKP_19', 'Increased ALT/GGT/AST_20', 'USG <1.20_21', 'Proteinuria_22']].info()
```

### Boolean data types

```python
#Replace yes & no values with true and false
data_df = data_df.replace({'Yes': True, 'No': False})

#Check df data types
data_df.info()
```

### Object data types

```python
#Change Case_1 to object data type
data_df['Case_1'] = data_df['Case_1'].astype(object)

#Check df
data_df.info()
```

### Drop NA Values

```python
#NAN Summary
data_df.isna().sum()
```

```python
#Removal of columns with >50% NAN
del data_df['Heart rate_23']
del data_df['Temperature_24']
del data_df['USG <1.20_21']
del data_df['Proteinuria_22']

#Check df
data_df.head()
```

### Rename columns

```python
#Collect list of column names
column_names = data_df.columns.to_list()

#Check list
column_names
```

```python
#Create dictionary of column headings & renames
col_names = {
    'Case_1': 'Diagnosis',
    'SourceBreed': 'Breed',
    'Age': 'Age',
    'Sex': 'Sex',
    'IsNeutered': 'Neutered',
    'Weight_11': 'WeightChange',
    'Weight at suspected': 'Weight',
    'Hospitalised': 'Hospitalised',
    'Vomiting_8': 'Vomiting',
    'Diarrhoea_9': 'Diarrhoea',
    'Potbelly/hepatomegaly_10': 'Hepatomegaly',
    'Thin/dry skin_12': 'ThinDrySkin',
    'Alopecia_13': 'Alopecia',
    'Pruritus_14': 'Pruritus',
    'Muscle wastage/weakness_15': 'MuscleWastageWeakness',
    'Lethargy_16': 'Lethargy',
    'Panting_17': 'Panting',
    'Neurological signs_18': 'Neurological',
    'Polydipsia mentioned_5': 'Polydipsia',
    'Polyuria mentioned_6': 'Polyuria',
    'Polyphagia_7': 'Polyphagia',
    'Increased ALKP_19': 'IncreasedALKP',
    'Increased ALT/GGT/AST_20': 'IncreasedATL_GGT_AST',
    'Hypertensive': 'Hypertensive',
    'Insulin': 'LowBloodSugar',
    'Hypothyroid': 'Hypothyroid',
    'Cruciate': 'CruciateSurgery'
    }

#Rename columns
clean_df = data_df.rename(columns=col_names)

#Reorder Columns
clean_df = clean_df[col_names.values()]

#Check df
clean_df
```

### Export CSV
